**61)Binary Search**

```python
li = [10, 20, 30, 40, 50]
key = 40
low = 0
high = len(li) - 1
while low <= high:
    mid = (low + high) // 2
    if key > li[mid]:
        low = mid + 1
    elif key < li[mid]:
        high = mid - 1
    else:
        print("Key found")
        break
else:
    print("Key not found")
```

**63)Selection sort**

```python
li = [64, 25, 12, 22, 11]
n = len(li)
for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if li[j] < li[min_index]:
            min_index = j
    li[i], li[min_index] = li[min_index], li[i]
print(li)
```

**64)Insertion sort**

```python
li = [12, 11, 13, 5, 6]
n = len(li)
for i in range(1, n):
    key = li[i]
    j = i - 1
    while j >= 0 and li[j] > key:
        li[j + 1] = li[j]
        j = j - 1
    li[j + 1] = key
print(li)
```

**70)Longest Consecutive sequence in a list**

```python
li = [100, 4, 200, 1, 3, 2]
nums = set(li)
longest = 0
for num in nums:
    if num - 1 not in nums:
        current = num
        length = 1
        while current + 1 in nums:
            current += 1
            length += 1
        longest = max(longest, length)
print(longest)
```

**93)Checking if brackets are balanced or not**

```python
s = input("Enter brackets: ")
stack = []
pairs = {
    ')': '(',
    ']': '[',
    '}': '{'
}
balanced = True
for ch in s:
    if ch in "([{":
        stack.append(ch)
    else:
        if len(stack) == 0:
            balanced = False
            break
        if stack[-1] != pairs[ch]:
            balanced = False
            break
        stack.pop()
if balanced and len(stack) == 0:
    print("Balanced")
else:
    print("Not Balanced")
```

**94)Implement a Stack using a list**

```python
stack = []
while True:
    print("\n1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Display")
    print("5. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        items = list(map(int, input("Enter element(s): ").split()))
        stack.extend(items)
        print("Element(s) pushed.")
    elif choice == 2:
        if len(stack) == 0:
            print("Stack Underflow")
        else:
            print("Popped Element:", stack.pop())
    elif choice == 3:
        if len(stack) == 0:
            print("Stack is Empty")
        else:
            print("Top Element:", stack[-1])
    elif choice == 4:
        if len(stack) == 0:
            print("Stack is Empty")
        else:
            print("Stack:", stack)
    elif choice == 5:
        print("Exiting...")
        break
    else: 
        print("Invalid Choice")
```

**95)Implement a Queue using a list**

```python
queue = []
while True:
    print("\n1. Enqueue")
    print("2. Dequeue")
    print("3. Front")
    print("4. Display")
    print("5. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        items = list(map(int, input("Enter element(s): ").split()))
        queue.extend(items)
        print("Element(s) Enqueued.")
    elif choice == 2:
        if len(queue) == 0:
            print("Queue Underflow")
        else:
            print("Dequeued Element:", queue.pop(0))
    elif choice == 3:
        if len(queue) == 0:
            print("Queue is Empty")
        else:
            print("Front Element:", queue[0])
    elif choice == 4:
        if len(queue) == 0:
            print("Queue is Empty")
        else:
            print("Queue:", queue)
    elif choice == 5:
        print("Exiting...")
        break
    else:
        print("Invalid Choice")
```

**99)Generate all permutations of a string**

```python
from itertools import permutations
s = input("Enter a string: ")
for p in permutations(s):
    print("".join(p))
```

**100)Generate all combinations of a string**

```python
from itertools import combinations
s = input("Enter a string: ")
for r in range(1, len(s) + 1):
    for c in combinations(s, r):
        print("".join(c))
```
